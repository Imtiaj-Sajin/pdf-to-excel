```python
def normalize_value(text):
    text = text.replace(",", "").strip()

    if text.startswith("("):
        return -float(text.strip("()"))

    if "%" in text:
        return float(text.replace("%", "")) / 100

    if "$" in text:
        return float(text.replace("$", ""))

    return float(text)
```

```python
import pdfplumber
import json
import pandas as pd

def extract_market_data(pdf_path, page_number=3):
    """
    Extracts the two 'Key Indicators' tables from the specific page.
    """
    data_results = {
        "current_quarter": [],
        "annual_trends": []
    }

    with pdfplumber.open(pdf_path) as pdf:
        # Pages are 0-indexed, so page 3 is index 2.
        # Adjust if your '23' implies physical page number vs index.
        # Based on your text, I will assume the page index is known.
        try:
            page = pdf.pages[page_number - 1]
        except IndexError:
            print(f"Page {page_number} not found.")
            return None

        # ---------------------------------------------------------
        # STEP 1: LOCATE THE ANCHOR
        # We look for "KEY INDICATORS" to define the start of our area
        # ---------------------------------------------------------
        text_instances = page.search("KEY INDICATORS")
        if not text_instances:
            print("Could not find 'KEY INDICATORS' anchor.")
            return None

        anchor_top = text_instances[0]["top"]

        # ---------------------------------------------------------
        # STEP 2: DEFINE CROP ZONES (Fixed Layout Strategy)
        # ---------------------------------------------------------
        # Based on your provided coordinate data:
        # Table 1 Header starts ~340, Ends ~460
        # Table 2 starts after that.

        page_width = page.width

        # Crop Area 1: Current Quarter Table
        # top: just below "KEY INDICATORS" (approx +20 units)
        # bottom: approx 470 (just after "Submarket" row)
        t1_crop = page.crop((0, anchor_top + 10, page_width, 470))

        # Crop Area 2: Annual Trends Table
        # top: starts where T1 ends (470)
        # bottom: approx 600 (to capture the second table)
        t2_crop = page.crop((0, 470, page_width, 600))

        # ---------------------------------------------------------
        # STEP 3: CONFIGURE EXTRACTION SETTINGS
        # ---------------------------------------------------------
        # vertical_strategy="text" is crucial here. It uses the white space
        # between columns to detect separators, which is safer than relying
        # on potentially invisible lines.
        table_settings = {
            "vertical_strategy": "text",
            "horizontal_strategy": "lines", # Rows are clearly separated by lines
            "snap_tolerance": 3,
            "join_tolerance": 3,
        }

        # ---------------------------------------------------------
        # STEP 4: EXTRACT TABLE 1 (Current Quarter)
        # ---------------------------------------------------------
        table_1 = t1_crop.extract_table(table_settings)

        if table_1:
            # Clean headers (merge multi-line headers like "Market Asking\nRent")
            headers = [h.replace('\n', ' ') if h else '' for h in table_1[0]]

            # The PDF extraction might create empty columns if alignment is weird.
            # Filter out purely None/Empty columns if necessary, but usually mapping is safer.

            for row in table_1[1:]:
                # Create a dictionary for this row
                # We filter out None values to avoid JSON errors
                clean_row = [cell.strip() if cell else "" for cell in row]

                # Basic validation: ensure row has data
                if any(clean_row):
                    row_dict = {}
                    for i, header in enumerate(headers):
                        if i < len(clean_row):
                            # Handle Duplicate header names or empty headers
                            key = header if header else f"Column_{i}"
                            row_dict[key] = clean_row[i]
                    data_results["current_quarter"].append(row_dict)

        # ---------------------------------------------------------
        # STEP 5: EXTRACT TABLE 2 (Annual Trends)
        # ---------------------------------------------------------
        table_2 = t2_crop.extract_table(table_settings)

        if table_2:
            headers_t2 = [h.replace('\n', ' ') if h else '' for h in table_2[0]]

            for row in table_2[1:]:
                clean_row = [cell.strip() if cell else "" for cell in row]
                if any(clean_row):
                    row_dict = {}
                    for i, header in enumerate(headers_t2):
                        if i < len(clean_row):
                            key = header if header else f"Column_{i}"
                            row_dict[key] = clean_row[i]
                    data_results["annual_trends"].append(row_dict)

    return data_results

# ==========================================
# EXECUTION
# ==========================================

file_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

# 1. Extract
json_data = extract_market_data(file_path, page_number=3)

# 2. Output JSON
print("--- JSON OUTPUT ---")
json_output = json.dumps(json_data, indent=4)
print(json_output)

# 3. VERIFICATION (Recheck)
print("\n--- VERIFICATION TABLES ---")
if json_data:
    if json_data["current_quarter"]:
        print("\nTable 1: Current Quarter Analysis")
        df1 = pd.DataFrame(json_data["current_quarter"])
        # Set index to the first column (e.g., Malls, Power Center) for better view
        if not df1.empty:
            print(df1.to_string())

    if json_data["annual_trends"]:
        print("\nTable 2: Annual Trends")
        df2 = pd.DataFrame(json_data["annual_trends"])
        if not df2.empty:
            print(df2.to_string())
```

```python
import pdfplumber
import json
import pandas as pd

def extract_fixed_layout(pdf_path, page_number=3):
    results = {}

    with pdfplumber.open(pdf_path) as pdf:
        # Access the specific page (0-indexed)
        try:
            page = pdf.pages[page_number - 1]
        except IndexError:
            return {"error": "Page number out of range"}

        # Define vertical dividers (X coordinates) shared by both tables.
        # Based on the grey lines in your screenshot, the columns align vertically.
        # These lines split the page into 8 distinct columns.
        common_vertical_lines = [160, 215, 275, 330, 395, 455, 515]

        # =========================================================
        # TABLE 1: CURRENT QUARTER
        # =========================================================
        # Top: 355 (Just below the header row "Current Quarter... RBA...")
        # Bottom: 470 (Includes the "Submarket" row)
        # Left/Right: 30 to 585 (Covers full width)
        t1_crop = page.crop((30, 355, 585, 470))

        t1_settings = {
            "vertical_strategy": "explicit",
            "explicit_vertical_lines": common_vertical_lines,
            "horizontal_strategy": "text", # Safely detects rows based on spacing
        }

        table_1_raw = t1_crop.extract_table(t1_settings)

        # Hardcoded headers ensure 100% consistency
        t1_headers = [
             "RBA", "Vacancy Rate", "Market Asking Rent",
            "Availability Rate", "Net Absorption SF", "Deliveries SF", "Under Construction"
        ]

        t1_data = []
        if table_1_raw:
            for row in table_1_raw:
                # Clean up each cell
                clean_row = [str(cell).strip() if cell else "-" for cell in row]

                # Filter out empty rows or purely divider rows
                if any(c != "-" and c != "None" for c in clean_row):
                    # Ensure row length matches headers
                    if len(clean_row) < len(t1_headers):
                         clean_row += ["-"] * (len(t1_headers) - len(clean_row))

                    row_dict = dict(zip(t1_headers, clean_row))
                    t1_data.append(row_dict)

        results["current_quarter"] = t1_data

        # =========================================================
        # TABLE 2: ANNUAL TRENDS
        # =========================================================
        # Top: 545 (Just below header row "Annual Trends... 12 Month...")
        # Bottom: 650 (Covers "Sales Volume" row)
        t2_crop = page.crop((30, 545, 585, 650))

        t2_settings = {
            "vertical_strategy": "explicit",
            "explicit_vertical_lines": common_vertical_lines,
            "horizontal_strategy": "text"
        }

        table_2_raw = t2_crop.extract_table(t2_settings)

        t2_headers = [
            "Metric", "12 Month", "Historical Average", "Forecast Average",
            "Peak", "When_Peak", "Trough", "When_Trough"
        ]

        t2_data = []
        if table_2_raw:
            for row in table_2_raw:
                clean_row = [str(cell).strip() if cell else "-" for cell in row]

                if any(c != "-" and c != "None" for c in clean_row):
                    if len(clean_row) < len(t2_headers):
                         clean_row += ["-"] * (len(t2_headers) - len(clean_row))

                    row_dict = dict(zip(t2_headers, clean_row))
                    t2_data.append(row_dict)

        results["annual_trends"] = t2_data

    return results

# ==========================================
# EXECUTION & VERIFICATION
# ==========================================

file_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

# 1. Run Extraction
data = extract_fixed_layout(file_path, page_number=3)

# 2. JSON Output
print("--- JSON RESULT ---")
print(json.dumps(data, indent=4))

# 3. Pandas Verification (To visualize table integrity)
print("\n--- VERIFICATION TABLES ---")
if "current_quarter" in data and data["current_quarter"]:
    print("\nTable 1: Current Quarter Analysis")
    df1 = pd.DataFrame(data["current_quarter"])
    print(df1.to_string(index=False))

if "annual_trends" in data and data["annual_trends"]:
    print("\nTable 2: Annual Trends")
    df2 = pd.DataFrame(data["annual_trends"])
    print(df2.to_string(index=False))
```

```python
import pdfplumber
from PIL import Image
import os

# CONFIGURATION
pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"
output_folder = "table_debug_images"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def debug_table_detection(pdf_path):
    print(f"Processing {pdf_path}...")

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1
            print(f"Scanning Page {page_num}...")

            # Convert page to image for visualization
            im = page.to_image(resolution=150)

            # 1. FIND TABLES
            # We use default settings first. This looks for solid lines.
            tables = page.find_tables(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines",
                "snap_tolerance": 3,
            })

            if tables:
                print(f"  -> Found {len(tables)} tables on Page {page_num}")

                # Draw Red Box around the whole table
                for table in tables:
                    bbox = table.bbox
                    im.draw_rect(bbox, stroke="red", stroke_width=3)

                    # Draw Blue Lines for rows/cols inside
                    # This helps us see if it's splitting cells correctly
                    for row in table.rows:
                        for cell in row.cells:
                            if cell: # cell is a tuple (x0, top, x1, bottom)
                                im.draw_rect(cell, stroke="blue", stroke_width=1)
            else:
                print(f"  -> No tables detected on Page {page_num} (using line detection).")

            # Save the debug image
            image_path = f"{output_folder}/page_{page_num}_debug.png"
            im.save(image_path)

    print(f"\nDone. Please check the folder '{output_folder}' and look at the images.")

debug_table_detection(pdf_path)
```

```python
import pdfplumber
import json
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def extract_all_tables(pdf_path):
    full_data = {}

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1

            # Detect tables
            tables = page.find_tables(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines",
                "snap_tolerance": 3,
            })

            if tables:
                page_key = f"Page_{page_num}"
                full_data[page_key] = []

                for idx, table in enumerate(tables):
                    # Extract the data
                    # extract() returns a list of lists e.g. [['Header', 'Val'], ['A', '1']]
                    table_data = table.extract()

                    # CLEANING:
                    # 1. Remove newlines inside cells
                    # 2. Convert None to empty string
                    cleaned_data = []
                    for row in table_data:
                        clean_row = [
                            str(cell).replace('\n', ' ').strip() if cell is not None else ""
                            for cell in row
                        ]
                        # Filter out completely empty rows
                        if any(clean_row):
                            cleaned_data.append(clean_row)

                    if cleaned_data:
                        full_data[page_key].append({
                            "table_id": idx + 1,
                            "data": cleaned_data
                        })

    return full_data

# 1. RUN EXTRACTION
json_result = extract_all_tables(pdf_path)

# 2. PRINT JSON
print("--- JSON OUTPUT ---")
print(json.dumps(json_result, indent=4))

# 3. VISUALIZATION (Verify with Pandas)
print("\n--- VISUALIZATION ---")
for page, tables in json_result.items():
    for table_obj in tables:
        print(f"\n=== {page} | Table {table_obj['table_id']} ===")
        rows = table_obj['data']

        if len(rows) > 0:
            # Assume first row is header for visualization
            headers = rows[0]
            data_rows = rows[1:]

            # Create DataFrame
            try:
                df = pd.DataFrame(data_rows, columns=headers)
                print(df.to_string())
            except Exception as e:
                # Fallback if columns don't match
                print(pd.DataFrame(rows).to_string(header=False))
        else:
            print("Empty Table")
```
