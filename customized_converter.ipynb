```python
def normalize_value(text):
    text = text.replace(",", "").strip()

    if text.startswith("("):
        return -float(text.strip("()"))

    if "%" in text:
        return float(text.replace("%", "")) / 100

    if "$" in text:
        return float(text.replace("$", ""))

    return float(text)
```

```python
import pdfplumber
import json
import pandas as pd

def extract_market_data(pdf_path, page_number=3):
    """
    Extracts the two 'Key Indicators' tables from the specific page.
    """
    data_results = {
        "current_quarter": [],
        "annual_trends": []
    }

    with pdfplumber.open(pdf_path) as pdf:
        # Pages are 0-indexed, so page 3 is index 2.
        # Adjust if your '23' implies physical page number vs index.
        # Based on your text, I will assume the page index is known.
        try:
            page = pdf.pages[page_number - 1]
        except IndexError:
            print(f"Page {page_number} not found.")
            return None

        # ---------------------------------------------------------
        # STEP 1: LOCATE THE ANCHOR
        # We look for "KEY INDICATORS" to define the start of our area
        # ---------------------------------------------------------
        text_instances = page.search("KEY INDICATORS")
        if not text_instances:
            print("Could not find 'KEY INDICATORS' anchor.")
            return None

        anchor_top = text_instances[0]["top"]

        # ---------------------------------------------------------
        # STEP 2: DEFINE CROP ZONES (Fixed Layout Strategy)
        # ---------------------------------------------------------
        # Based on your provided coordinate data:
        # Table 1 Header starts ~340, Ends ~460
        # Table 2 starts after that.

        page_width = page.width

        # Crop Area 1: Current Quarter Table
        # top: just below "KEY INDICATORS" (approx +20 units)
        # bottom: approx 470 (just after "Submarket" row)
        t1_crop = page.crop((0, anchor_top + 10, page_width, 470))

        # Crop Area 2: Annual Trends Table
        # top: starts where T1 ends (470)
        # bottom: approx 600 (to capture the second table)
        t2_crop = page.crop((0, 470, page_width, 600))

        # ---------------------------------------------------------
        # STEP 3: CONFIGURE EXTRACTION SETTINGS
        # ---------------------------------------------------------
        # vertical_strategy="text" is crucial here. It uses the white space
        # between columns to detect separators, which is safer than relying
        # on potentially invisible lines.
        table_settings = {
            "vertical_strategy": "text",
            "horizontal_strategy": "lines", # Rows are clearly separated by lines
            "snap_tolerance": 3,
            "join_tolerance": 3,
        }

        # ---------------------------------------------------------
        # STEP 4: EXTRACT TABLE 1 (Current Quarter)
        # ---------------------------------------------------------
        table_1 = t1_crop.extract_table(table_settings)

        if table_1:
            # Clean headers (merge multi-line headers like "Market Asking\nRent")
            headers = [h.replace('\n', ' ') if h else '' for h in table_1[0]]

            # The PDF extraction might create empty columns if alignment is weird.
            # Filter out purely None/Empty columns if necessary, but usually mapping is safer.

            for row in table_1[1:]:
                # Create a dictionary for this row
                # We filter out None values to avoid JSON errors
                clean_row = [cell.strip() if cell else "" for cell in row]

                # Basic validation: ensure row has data
                if any(clean_row):
                    row_dict = {}
                    for i, header in enumerate(headers):
                        if i < len(clean_row):
                            # Handle Duplicate header names or empty headers
                            key = header if header else f"Column_{i}"
                            row_dict[key] = clean_row[i]
                    data_results["current_quarter"].append(row_dict)

        # ---------------------------------------------------------
        # STEP 5: EXTRACT TABLE 2 (Annual Trends)
        # ---------------------------------------------------------
        table_2 = t2_crop.extract_table(table_settings)

        if table_2:
            headers_t2 = [h.replace('\n', ' ') if h else '' for h in table_2[0]]

            for row in table_2[1:]:
                clean_row = [cell.strip() if cell else "" for cell in row]
                if any(clean_row):
                    row_dict = {}
                    for i, header in enumerate(headers_t2):
                        if i < len(clean_row):
                            key = header if header else f"Column_{i}"
                            row_dict[key] = clean_row[i]
                    data_results["annual_trends"].append(row_dict)

    return data_results

# ==========================================
# EXECUTION
# ==========================================

file_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

# 1. Extract
json_data = extract_market_data(file_path, page_number=3)

# 2. Output JSON
print("--- JSON OUTPUT ---")
json_output = json.dumps(json_data, indent=4)
print(json_output)

# 3. VERIFICATION (Recheck)
print("\n--- VERIFICATION TABLES ---")
if json_data:
    if json_data["current_quarter"]:
        print("\nTable 1: Current Quarter Analysis")
        df1 = pd.DataFrame(json_data["current_quarter"])
        # Set index to the first column (e.g., Malls, Power Center) for better view
        if not df1.empty:
            print(df1.to_string())

    if json_data["annual_trends"]:
        print("\nTable 2: Annual Trends")
        df2 = pd.DataFrame(json_data["annual_trends"])
        if not df2.empty:
            print(df2.to_string())
```

```python
import pdfplumber
import json
import pandas as pd

def extract_fixed_layout(pdf_path, page_number=3):
    results = {}

    with pdfplumber.open(pdf_path) as pdf:
        # Access the specific page (0-indexed)
        try:
            page = pdf.pages[page_number - 1]
        except IndexError:
            return {"error": "Page number out of range"}

        # Define vertical dividers (X coordinates) shared by both tables.
        # Based on the grey lines in your screenshot, the columns align vertically.
        # These lines split the page into 8 distinct columns.
        common_vertical_lines = [160, 215, 275, 330, 395, 455, 515]

        # =========================================================
        # TABLE 1: CURRENT QUARTER
        # =========================================================
        # Top: 355 (Just below the header row "Current Quarter... RBA...")
        # Bottom: 470 (Includes the "Submarket" row)
        # Left/Right: 30 to 585 (Covers full width)
        t1_crop = page.crop((30, 355, 585, 470))

        t1_settings = {
            "vertical_strategy": "explicit",
            "explicit_vertical_lines": common_vertical_lines,
            "horizontal_strategy": "text", # Safely detects rows based on spacing
        }

        table_1_raw = t1_crop.extract_table(t1_settings)

        # Hardcoded headers ensure 100% consistency
        t1_headers = [
             "RBA", "Vacancy Rate", "Market Asking Rent",
            "Availability Rate", "Net Absorption SF", "Deliveries SF", "Under Construction"
        ]

        t1_data = []
        if table_1_raw:
            for row in table_1_raw:
                # Clean up each cell
                clean_row = [str(cell).strip() if cell else "-" for cell in row]

                # Filter out empty rows or purely divider rows
                if any(c != "-" and c != "None" for c in clean_row):
                    # Ensure row length matches headers
                    if len(clean_row) < len(t1_headers):
                         clean_row += ["-"] * (len(t1_headers) - len(clean_row))

                    row_dict = dict(zip(t1_headers, clean_row))
                    t1_data.append(row_dict)

        results["current_quarter"] = t1_data

        # =========================================================
        # TABLE 2: ANNUAL TRENDS
        # =========================================================
        # Top: 545 (Just below header row "Annual Trends... 12 Month...")
        # Bottom: 650 (Covers "Sales Volume" row)
        t2_crop = page.crop((30, 545, 585, 650))

        t2_settings = {
            "vertical_strategy": "explicit",
            "explicit_vertical_lines": common_vertical_lines,
            "horizontal_strategy": "text"
        }

        table_2_raw = t2_crop.extract_table(t2_settings)

        t2_headers = [
            "Metric", "12 Month", "Historical Average", "Forecast Average",
            "Peak", "When_Peak", "Trough", "When_Trough"
        ]

        t2_data = []
        if table_2_raw:
            for row in table_2_raw:
                clean_row = [str(cell).strip() if cell else "-" for cell in row]

                if any(c != "-" and c != "None" for c in clean_row):
                    if len(clean_row) < len(t2_headers):
                         clean_row += ["-"] * (len(t2_headers) - len(clean_row))

                    row_dict = dict(zip(t2_headers, clean_row))
                    t2_data.append(row_dict)

        results["annual_trends"] = t2_data

    return results

# ==========================================
# EXECUTION & VERIFICATION
# ==========================================

file_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

# 1. Run Extraction
data = extract_fixed_layout(file_path, page_number=3)

# 2. JSON Output
print("--- JSON RESULT ---")
print(json.dumps(data, indent=4))

# 3. Pandas Verification (To visualize table integrity)
print("\n--- VERIFICATION TABLES ---")
if "current_quarter" in data and data["current_quarter"]:
    print("\nTable 1: Current Quarter Analysis")
    df1 = pd.DataFrame(data["current_quarter"])
    print(df1.to_string(index=False))

if "annual_trends" in data and data["annual_trends"]:
    print("\nTable 2: Annual Trends")
    df2 = pd.DataFrame(data["annual_trends"])
    print(df2.to_string(index=False))
```

```python
import pdfplumber
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def extract_table_with_crop(pdf_path):
    full_data = []

    # The specific header text that signals the start of this table
    TARGET_HEADER = "SALE COMPARABLES SUMMARY STATISTICS"

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):

            # 1. SEARCH for the table header text
            # We look for the specific title to know where to start cropping
            found_header = page.search(TARGET_HEADER)

            if found_header:
                # Get the bounding box of the text: [x0, top, x1, bottom]
                header_bbox = found_header[0]
                header_bottom_y = header_bbox['bottom']

                # 2. CROP the page
                # We crop from the bottom of the header to the bottom of the page
                # (You can adjust 'page.height' if there is a footer you want to avoid)
                crop_box = (0, header_bottom_y, page.width, page.height)
                cropped_page = page.crop(crop_box)

                # 3. EXTRACT from the cropped area
                # vertical_strategy="text" works perfectly now because the map is gone
                tables = cropped_page.find_tables(table_settings={
                    "vertical_strategy": "text",      # Use text alignment for columns
                    "horizontal_strategy": "lines",   # Use lines for rows
                    "intersection_y_tolerance": 5,    # Help merge close lines
                    "snap_tolerance": 3,
                })

                for idx, table in enumerate(tables):
                    data = table.extract()

                    # Basic cleaning
                    cleaned_data = []
                    for row in data:
                        # Remove newlines and None
                        clean_row = [str(cell).replace('\n', ' ').strip() if cell else "" for cell in row]
                        # Only add if row has data
                        if any(clean_row):
                            cleaned_data.append(clean_row)

                    if cleaned_data:
                        full_data.append({
                            "page": i + 1,
                            "table_id": idx + 1,
                            "data": cleaned_data
                        })

    return full_data

# --- RUN ---
results = extract_table_with_crop(pdf_path)

# --- VISUALIZE ---
for item in results:
    print(f"\n=== Page {item['page']} Table {item['table_id']} ===")
    if item['data']:
        df = pd.DataFrame(item['data'][1:], columns=item['data'][0])
        print(df.to_string())
```

```python
import pdfplumber
import json
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def clean_row_data(row):
    """Helper to clean cell data."""
    return [
        str(cell).replace('\n', ' ').strip() if cell is not None else ""
        for cell in row
    ]

def extract_with_crop_and_text(page, header_text):
    """
    Strategy for tables missing side borders (like Summary Statistics).
    1. Finds header.
    2. Crops page below header.
    3. Uses whitespace (text) to define columns.
    """
    extracted_tables = []

    # Find the header location
    found_header = page.search(header_text)

    if found_header:
        # Get position of the header
        bbox = found_header[0]

        # Crop from bottom of header to end of page
        # Adjust 'top' slightly (+1) to avoid catching the header underline
        crop_box = (0, bbox['bottom'] + 2, page.width, page.height)
        cropped_page = page.crop(crop_box)

        # settings: vertical_strategy="text" is key here
        tables = cropped_page.find_tables(table_settings={
            "vertical_strategy": "text",
            "horizontal_strategy": "lines",
            "snap_tolerance": 4,
            "intersection_y_tolerance": 5,
        })

        for table in tables:
            data = table.extract()
            cleaned_data = [clean_row_data(row) for row in data if any(clean_row_data(row))]
            if cleaned_data:
                extracted_tables.append(cleaned_data)

    return extracted_tables

def extract_standard_grid(page):
    """
    Strategy for standard tables with drawn borders (Market Stats, Property Lists).
    Uses 'lines' to define columns.
    """
    extracted_tables = []

    # settings: vertical_strategy="lines" is safer for big property lists
    tables = page.find_tables(table_settings={
        "vertical_strategy": "lines",
        "horizontal_strategy": "lines",
        "snap_tolerance": 4,
        "join_tolerance": 3,
        "edge_min_length": 3,
        "min_words_vertical": 1,
        "min_words_horizontal": 1,
    })

    for table in tables:
        data = table.extract()
        cleaned_data = [clean_row_data(row) for row in data if any(clean_row_data(row))]
        if cleaned_data:
            extracted_tables.append(cleaned_data)

    return extracted_tables

def extract_all_tables_robust(pdf_path):
    full_data = {}

    # The trigger text to switch strategies
    SPECIAL_HEADER_TRIGGER = "OVERALL RENT & VACANCY"

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1
            page_key = f"Page_{page_num}"
            full_data[page_key] = []

            # 1. CHECK STRATEGY
            # Check if this page contains the problematic table
            page_text = page.extract_text() or ""

            if SPECIAL_HEADER_TRIGGER in page_text:
                # USE STRATEGY A: Crop + Text
                print(f"Processing Page {page_num}: Detected Summary Stats (Using Text Strategy)")
                tables_data = extract_with_crop_and_text(page, SPECIAL_HEADER_TRIGGER)
            else:
                # USE STRATEGY B: Standard Lines
                # print(f"Processing Page {page_num}: Standard Table (Using Lines Strategy)")
                tables_data = extract_standard_grid(page)

            # 2. STORE DATA
            for idx, t_data in enumerate(tables_data):
                full_data[page_key].append({
                    "table_id": idx + 1,
                    "data": t_data
                })

    return full_data

# --- 1. EXECUTE ---
json_result = extract_all_tables_robust(pdf_path)

# --- 2. JSON OUTPUT (Optional) ---
# print(json.dumps(json_result, indent=4))

# --- 3. VERIFY WITH PANDAS ---
print("\n--- VISUALIZATION ---")
for page, tables in json_result.items():
    for table_obj in tables:
        print(f"\n=== {page} | Table {table_obj['table_id']} ===")
        rows = table_obj['data']

        if len(rows) > 0:
            headers = rows[0]
            data_rows = rows[1:]

            # Attempt to create a clean DataFrame
            try:
                # Handle cases where data row length mismatches header length
                df = pd.DataFrame(data_rows, columns=headers)
                print(df.to_string())
            except Exception:
                # Fallback print if columns don't match perfectly
                print(pd.DataFrame(rows).to_string(header=False))
        else:
            print("Empty Table")
```

```python
import pdfplumber
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def clean_row(row):
    """Cleans newline characters and whitespace from a row."""
    return [str(cell).replace('\n', ' ').strip() if cell is not None else "" for cell in row]

def extract_table_from_crop(page, crop_box):
    """
    Extracts a table from a specific cropped area using the text-strategy
    (most robust for missing borders).
    """
    try:
        cropped_page = page.crop(crop_box)
        # Use 'text' strategy for columns to catch floating data (like "High" column)
        # Use 'lines' strategy for rows as they are clearly drawn
        tables = cropped_page.find_tables(table_settings={
            "vertical_strategy": "text",
            "horizontal_strategy": "lines",
            "snap_tolerance": 4,
            "intersection_y_tolerance": 5,
        })

        if tables:
            # Return the data of the first table found in this crop
            data = tables[0].extract()
            return [clean_row(row) for row in data if any(clean_row(row))]
    except Exception as e:
        print(f"Error extracting from crop: {e}")
    return None

def extract_targeted_tables(pdf_path):
    extracted_data = {}

    # Configuration
    TARGET_HEADER = "OVERALL RENT & VACANCY"
    global_table_counter = 0

    # We will track if we found the header table to avoid duplication if it overlaps
    header_table_found = False

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1

            # --- STRATEGY A: HUNT FOR SPECIFIC HEADER ---
            # Search for "OVERALL RENT & VACANCY"
            if not header_table_found:
                matches = page.search(TARGET_HEADER)
                if matches:
                    print(f"Found '{TARGET_HEADER}' on Page {page_num}")
                    header_bbox = matches[0] # Get the first match

                    # Crop from the text down to the bottom of the page
                    crop_box = (0, header_bbox['bottom'], page.width, page.height)

                    table_data = extract_table_from_crop(page, crop_box)
                    if table_data:
                        extracted_data["Target_Named_Table"] = {
                            "title": TARGET_HEADER,
                            "page": page_num,
                            "data": table_data
                        }
                        header_table_found = True

            # --- STRATEGY B: GRAB FIRST 2 TABLES IN DOC ---
            # We only look for tables if we haven't found the first 2 yet
            if global_table_counter < 2:
                # Find tables using 'lines' to locate them physically
                tables_on_page = page.find_tables(table_settings={
                    "vertical_strategy": "lines",
                    "horizontal_strategy": "lines"
                })

                # Sort by vertical position (top to bottom)
                tables_on_page.sort(key=lambda t: t.bbox[1])

                for table in tables_on_page:
                    if global_table_counter >= 2:
                        break

                    global_table_counter += 1

                    # Define Crop Box (Strip across full width to be robust)
                    y0, y1 = table.bbox[1], table.bbox[3]
                    crop_box = (0, y0, page.width, y1)

                    # Extract
                    table_data = extract_table_from_crop(page, crop_box)

                    if table_data:
                        key = f"Table_{global_table_counter}"
                        extracted_data[key] = {
                            "title": f"Doc Table #{global_table_counter}",
                            "page": page_num,
                            "data": table_data
                        }

    return extracted_data

# 1. RUN
results = extract_targeted_tables(pdf_path)

# 2. VISUALIZE
print("\n--- FINAL EXTRACTION RESULTS ---")

# Print Table 1
if "Table_1" in results:
    t = results["Table_1"]
    print(f"\n=== {t['title']} (Page {t['page']}) ===")
    df = pd.DataFrame(t['data'][1:], columns=t['data'][0])
    print(df.to_string())

# Print Table 2
if "Table_2" in results:
    t = results["Table_2"]
    print(f"\n=== {t['title']} (Page {t['page']}) ===")
    df = pd.DataFrame(t['data'][1:], columns=t['data'][0])
    print(df.to_string())

# Print Named Table
if "Target_Named_Table" in results:
    t = results["Target_Named_Table"]
    print(f"\n=== {t['title']} (Page {t['page']}) ===")
    df = pd.DataFrame(t['data'][1:], columns=t['data'][0])
    print(df.to_string())
else:
    print(f"\n=== Target Table 'OVERALL RENT & VACANCY' Not Found ===")
```

```python
import pdfplumber
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def clean_row(row):
    """Cleans newline characters and whitespace."""
    return [str(cell).replace('\n', ' ').strip() if cell is not None else "" for cell in row]

def extract_hybrid_table(page, y0, y1, header_row_count=1):
    """
    PERFECT HYBRID EXTRACTION:
    1. Crops the specific table area.
    2. Extracts HEADERS using 'lines' strategy (Fixes twisted text).
    3. Extracts DATA using 'text' strategy (Fixes missing columns like 'Year').
    4. Stitches them together.
    """
    try:
        # 1. Define Crop
        # Add small buffer to y-bounds to ensure borders are caught
        crop_box = (0, y0 - 2, page.width, y1 + 2)
        cropped_page = page.crop(crop_box)

        # 2. Extract Headers (Using LINES strategy for clean text)
        table_lines = cropped_page.find_tables(table_settings={
            "vertical_strategy": "lines",
            "horizontal_strategy": "lines",
            "snap_tolerance": 4,
        })

        headers = []
        if table_lines:
            # Extract only the top N rows for headers
            raw_lines = table_lines[0].extract()
            if raw_lines:
                headers = [clean_row(r) for r in raw_lines[:header_row_count]]

        # 3. Extract Data (Using TEXT strategy for robust columns)
        table_text = cropped_page.find_tables(table_settings={
            "vertical_strategy": "text",
            "horizontal_strategy": "lines",
            "snap_tolerance": 4,
            "intersection_y_tolerance": 5,
        })

        data_rows = []
        if table_text:
            # Sort by area to ensure we get the main table
            table_text.sort(key=lambda x: (x.bbox[2]-x.bbox[0]) * (x.bbox[3]-x.bbox[1]), reverse=True)
            raw_text = table_text[0].extract()

            # Skip the header rows from the text extraction (because they are twisted)
            # and keep the data rows
            if len(raw_text) > header_row_count:
                data_rows = [clean_row(r) for r in raw_text[header_row_count:]]

        # 4. Stitch and Validate
        # If we failed to get clean headers, fall back to text headers
        if not headers and data_rows:
            return data_rows

        # Filter empty data rows
        clean_data = [r for r in data_rows if any(r)]

        return headers + clean_data

    except Exception as e:
        print(f"Hybrid Extraction Error: {e}")
        return None

def extract_final_perfect(pdf_path):
    results = {}
    tables_found_count = 0
    named_table_found = False
    target_header = "OVERALL RENT & VACANCY"

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1

            # Locate physical tables using lines (best for coordinates)
            physical_tables = page.find_tables(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines"
            })
            physical_tables.sort(key=lambda t: t.bbox[1])

            # --- TASK 1: EXTRACT TABLE 1 & 2 ---
            if tables_found_count < 2 and physical_tables:
                for table in physical_tables:
                    if tables_found_count >= 2:
                        break

                    y0, y1 = table.bbox[1], table.bbox[3]

                    # Standard tables usually have 1 header row
                    data = extract_hybrid_table(page, y0, y1, header_row_count=1)

                    if data:
                        tables_found_count += 1
                        results[f"Table_{tables_found_count}"] = {
                            "page": page_num,
                            "data": data,
                            "title": f"Table {tables_found_count}"
                        }

            # --- TASK 2: EXTRACT 'OVERALL RENT & VACANCY' ---
            if not named_table_found:
                matches = page.search(target_header)
                if matches:
                    header_bottom = matches[0]['bottom']
                    print(f"Found '{target_header}' on Page {page_num}")

                    # Find first table below header
                    candidate_tables = [t for t in physical_tables if t.bbox[1] > header_bottom]

                    if candidate_tables:
                        target_table = candidate_tables[0]
                        y0, y1 = target_table.bbox[1], target_table.bbox[3]

                        # "Rent & Vacancy" table has 2 header rows (Super header + Sub header)
                        data = extract_hybrid_table(page, y0, y1, header_row_count=2)

                        if data:
                            results["Named_Table"] = {
                                "title": target_header,
                                "page": page_num,
                                "data": data
                            }
                            named_table_found = True

    return results

# --- EXECUTE ---
final_data = extract_final_perfect(pdf_path)

# --- VISUALIZE ---
print("\n" + "="*40)
print(" PERFECT HYBRID OUTPUT ")
print("="*40)

keys = ["Table_1", "Table_2", "Named_Table"]

for key in keys:
    if key in final_data:
        item = final_data[key]
        print(f"\n[{item['title']}] - Page {item['page']}")

        rows = item['data']
        if rows:
            # Logic to create pretty DataFrame
            # If multiple header rows, we join them or just take the last one for column names
            try:
                if len(rows) > 1:
                    # If it's the Named Table, it has 2 header rows
                    if key == "Named_Table":
                        # Use Row 2 (Index 1) as the primary columns
                        cols = rows[1]
                        # Data starts at Row 3 (Index 2)
                        df = pd.DataFrame(rows[2:], columns=cols)
                    else:
                        # Standard Table
                        cols = rows[0]
                        df = pd.DataFrame(rows[1:], columns=cols)

                    print(df.to_string())
                else:
                    print(rows)
            except Exception as e:
                # Fallback if column counts mismatch between Line-Header and Text-Data
                print("Warning: Column count mismatch between Header and Data. Printing Raw List:")
                for r in rows:
                    print(r)
        else:
            print("Empty Data")
    else:
        print(f"\n[{key}] Not Found")
```

```python
import pdfplumber
import pandas as pd

pdf_path = "/content/Goddard-West-Retail-Submarket-2024-10-17.pdf"

def clean_row(row):
    """Cleans newline characters and whitespace."""
    return [str(cell).replace('\n', ' ').strip() if cell is not None else "" for cell in row]

def extract_standard_table(page, crop_box):
    """
    Standard extraction using LINES.
    Best for Table 1 & 2 to ensure clean headers.
    """
    try:
        cropped_page = page.crop(crop_box)
        tables = cropped_page.find_tables(table_settings={
            "vertical_strategy": "lines",
            "horizontal_strategy": "lines",
            "snap_tolerance": 4,
        })
        if tables:
            data = tables[0].extract()
            return [clean_row(row) for row in data if any(clean_row(row))]
    except Exception:
        pass
    return None

def extract_hybrid_named_table(page, crop_box):
    """
    Hybrid extraction specifically for 'OVERALL RENT & VACANCY'.
    - Uses LINES for Headers (Rows 0-1) to avoid twisted text.
    - Uses TEXT for Data (Rows 2+) to capture the 'Year' column.
    """
    try:
        cropped_page = page.crop(crop_box)

        # 1. Get Headers using LINES strategy (Cleanest)
        header_tables = cropped_page.find_tables(table_settings={
            "vertical_strategy": "lines",
            "horizontal_strategy": "lines",
        })
        headers = []
        if header_tables:
            # We assume the first 2 rows are headers for this specific table
            raw_lines = header_tables[0].extract()
            if len(raw_lines) >= 2:
                headers = [clean_row(r) for r in raw_lines[:2]]

        # 2. Get Data using TEXT strategy (Robust for missing borders)
        data_tables = cropped_page.find_tables(table_settings={
            "vertical_strategy": "text",
            "horizontal_strategy": "lines",
        })
        data_rows = []
        if data_tables:
            # Sort to ensure we get the main table
            data_tables.sort(key=lambda x: (x.bbox[2]-x.bbox[0]) * (x.bbox[3]-x.bbox[1]), reverse=True)
            raw_text = data_tables[0].extract()

            # Skip the first 2 rows (headers) and take the rest as data
            if len(raw_text) > 2:
                data_rows = [clean_row(r) for r in raw_text[2:]]

        # 3. Stitch them together
        if headers and data_rows:
            return headers + data_rows

    except Exception as e:
        print(f"Hybrid Error: {e}")
    return None

def extract_final_tables(pdf_path):
    results = {}
    tables_found_count = 0
    named_table_found = False
    target_header = "OVERALL RENT & VACANCY"

    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages):
            page_num = i + 1

            # Locate physical tables using lines
            physical_tables = page.find_tables(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines"
            })
            physical_tables.sort(key=lambda t: t.bbox[1])

            # --- TASK 1: TABLE 1 & 2 (STANDARD STRATEGY) ---
            if tables_found_count < 2 and physical_tables:
                for table in physical_tables:
                    if tables_found_count >= 2:
                        break

                    y0, y1 = table.bbox[1], table.bbox[3]
                    crop_box = (0, y0 - 2, page.width, y1 + 2)

                    # Use STANDARD extraction for these (Clean headers)
                    data = extract_standard_table(page, crop_box)

                    if data:
                        tables_found_count += 1
                        results[f"Table_{tables_found_count}"] = {
                            "page": page_num,
                            "data": data,
                            "title": f"Table {tables_found_count}"
                        }

            # --- TASK 2: NAMED TABLE (HYBRID STRATEGY) ---
            if not named_table_found:
                matches = page.search(target_header)
                if matches:
                    header_bottom = matches[0]['bottom']
                    print(f"Found '{target_header}' on Page {page_num}")

                    # Find first table below header
                    candidate_tables = [t for t in physical_tables if t.bbox[1] > header_bottom]

                    if candidate_tables:
                        target_table = candidate_tables[0]
                        y0, y1 = target_table.bbox[1], target_table.bbox[3]
                        crop_box = (0, y0 - 2, page.width, y1 + 2)

                        # Use HYBRID extraction for this specific table
                        data = extract_hybrid_named_table(page, crop_box)

                        if data:
                            results["Named_Table"] = {
                                "title": target_header,
                                "page": page_num,
                                "data": data
                            }
                            named_table_found = True

    return results

# --- EXECUTE ---
final_data = extract_final_tables(pdf_path)

# --- VISUALIZE ---
print("\n" + "="*40)
print(" FINAL OUTPUT ")
print("="*40)

keys = ["Table_1", "Table_2", "Named_Table"]

for key in keys:
    if key in final_data:
        item = final_data[key]
        print(f"\n[{item['title']}] - Page {item['page']}")

        rows = item['data']
        if rows:
            try:
                # Pretty print logic
                if key == "Named_Table" and len(rows) > 1:
                    # Named Table: Row 0 = Super Header, Row 1 = Sub Header
                    # We print using Row 1 as columns
                    df = pd.DataFrame(rows[2:], columns=rows[1])
                    print(df.to_string())
                else:
                    # Standard Tables
                    df = pd.DataFrame(rows[1:], columns=rows[0])
                    print(df.to_string())
            except Exception:
                 print(pd.DataFrame(rows).to_string(header=False))
        else:
            print("Empty Data")
    else:
        print(f"\n[{key}] Not Found")
```

```python
import pdfplumber
import pandas as pd

pdf_path = "/content/Properties _ 6 McLaren.pdf"

def extract_demographic_frames(pdf_path):
    print(f"--- PROCESSING: {pdf_path.split('/')[-1]} ---")

    # We want these specific tables
    TARGET_TITLES = ["Population", "Population By Race", "Housing", "Households", "Income"]
    captured_count = 0

    with pdfplumber.open(pdf_path) as pdf:
        for p_index, page in enumerate(pdf.pages):
            print(f"\nScanning Page {p_index + 1}...")

            # STEP 1: DETECT OUTER FRAMES
            # We look for the solid lines that box these tables
            frames = page.find_tables(table_settings={
                "vertical_strategy": "lines",
                "horizontal_strategy": "lines",
                "snap_tolerance": 5,
            })

            # Filter out small artifacts, keep only substantial tables (width > 200)
            valid_frames = [f for f in frames if (f.bbox[2] - f.bbox[0]) > 200]

            # Sort by vertical position
            valid_frames.sort(key=lambda f: f.bbox[1])

            print(f"-> Found {len(valid_frames)} framed areas.")

            for i, frame in enumerate(valid_frames):
                if captured_count >= 5:
                    print("\n[STOPPING: Found 5 Tables]")
                    return

                # STEP 2: CROP THE FRAME
                # Get the coordinates of the box
                bbox = frame.bbox
                try:
                    cropped_frame = page.crop(bbox)
                except:
                    continue

                # Check text to see if it's a target table
                text = cropped_frame.extract_text()
                if not text:
                    continue

                # Identify which table this is based on header text
                table_title = "Unknown"
                is_target = False

                # Check for "3 miles" signature which signifies a data table
                if "3 miles" in text or "5 miles" in text:
                    is_target = True

                # Try to find the specific name
                first_line = text.split('\n')[0]
                for title in TARGET_TITLES:
                    if title in first_line:
                        table_title = title
                        is_target = True
                        break

                if is_target:
                    captured_count += 1
                    print(f"\n=== Table {captured_count}: {table_title} (Page {p_index + 1}) ===")

                    # STEP 3: EXTRACT CONTENT USING TEXT STRATEGY
                    # We use looser tolerance to prevent word splitting
                    content_tables = cropped_frame.find_tables(table_settings={
                        "vertical_strategy": "text",
                        "horizontal_strategy": "text",
                        "snap_tolerance": 4,
                        "text_x_tolerance": 15,    # <--- Increased to keep "Population" together
                        "text_y_tolerance": 5,
                        "intersection_x_tolerance": 15
                    })

                    if content_tables:
                        # We extract the first table found in this crop
                        raw_data = content_tables[0].extract()

                        # Clean Data
                        clean_data = []
                        for row in raw_data:
                            # Clean cells
                            r = [str(cell).replace('\n', ' ').strip() if cell else "" for cell in row]
                            # Remove empty rows
                            if any(r):
                                clean_data.append(r)

                        # Print
                        if clean_data:
                            try:
                                # Use first row as header if it looks like header
                                df = pd.DataFrame(clean_data[1:], columns=clean_data[0])
                                print(df.to_string())
                            except:
                                print(pd.DataFrame(clean_data).to_string(header=False))
                        else:
                            print("[Empty Data]")

# Run the extraction
extract_demographic_frames(pdf_path)
```
